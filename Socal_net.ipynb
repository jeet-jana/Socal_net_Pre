```python
import seaborn as sns

data = pd.read_csv("Social_Network_Ads.csv")
data.head()

```

```python
sns.heatmap(data.isnull())
```

```python
data.drop(["User ID"], axis=1, inplace = True)
```

```python
data.Gender = [1 if value == "Male" else 0 for value in data.Gender]
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
x = data.drop(["Purchased"],axis = 1)
y = data.Purchased
Scaller = StandardScaler()

x_scaled = Scaller.fit_transform(x)
x_scaled
```

```python
data.Purchased.value_counts().plot(kind = "bar")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size = 0.30,random_state = 64)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,y_pred)

print(f"Accuracy: {accuracy}")
```
